### 교재 없는 내용 - openpyxl 
### 위의 예제 처럼 'xlsxwriter' 를 사용 해서 문서 혹은 시트를 새로 작성하면, 새로운 문서를 만들거나, 이름이 동일하면, 기존 내용을 지우고 덮어쓰게 된다.
### 기존 문서에 추가로 데이터를 기록 하려면, 'xlsxwriter'가 아닌 'openpyxl' 을 사용해야 하는데, 아래는 그 예제 코드이다.

```python
import pandas as pd
import openpyxl
from IPython.display import display

folder="C:/myPyExcel/data/ch05/"
wFile=folder+'사원별_월간_판매현황_two_sheets2.xlsx'

# load_workbook() 으로 wFile 을 불러와서 wb를 초기화 
wb=openpyxl.load_workbook(wFile)

# 추가하려는 딕셔너리 데이터
first=[
    {'이름':'홍길동', '1월':78,'2월':35, '3월':52,'4월':73, '5월':43, '6월':59},
    {'이름':'강호동', '1월':45,'2월':37, '3월':18,'4월':65, '5월':39, '6월':74}   
    ]

second=[
    {'이름':'신동엽', '7월':78,'8월':35, '9월':52,'10월':73, '11월':43, '12월':59},
    {'이름':'유재석','7월':63,'8월':41, '9월':28,'10월':37, '11월':55, '12월':73}   
    ]

#추가 데이터를 DataFrame으로 변환
df3=pd.DataFrame(first)
df4=pd.DataFrame(second)

# 엑셀에 기록하기위한 Pandas의 ExcelWriter 객체 writer 를 , openpyxl 을 사용하여 생성
with pd.ExcelWriter(wFile,engine='openpyxl') as writer:
    writer.book=wb # writer의 book에 wb 저장    
    # wirter의 sheets 를 초기화 하여 저장 하는거 같은데, 여기를 모르겠음
    # 아래 writer.sheets={ }  이 부분을 주석 처리하면 아래 기록 하는 곳에서 '상반기1', '하반기1 이라는 새로운 시트를 만들어 버림
   
    writer.sheets={
    ws.title:ws
    for ws in wb.worksheets
    }   

    df3.to_excel(writer,sheet_name='상반기', startrow=wb.worksheets[0].max_row, index=False, header=False)
    df4.to_excel(writer,sheet_name='하반기', startrow=wb.worksheets[1].max_row, index=False, header=False)

display(df3)
display(df4)
display(writer.book)
```

### Pandas의 표 데이터 연산
#### 두개의 데이터를 단순히 사칙연산 해 준다.
- series + 상수 :  series의 모든 값에 상수를 더해서 반환
- series + series : Numpy에서는 서로 길이가 다른 두개의 배열은 연산 할수 없지만, Series는 가능하다
    - 두개의 시리즈 같은 위치에 있는 값 끼리 덧셈을 진행 한다. 
    - 두 개의 시리즈는 길이가 달라도 연산이 가능하며, 이 경우 연산 할 수 있는 요소 끼리만 연산을 수행 하고, 연산 할 수 없는 요소는 NaN 으로 표시    
    <br>
- DataFrame + DataFrame : DataFrame 도 Series와 마찬가지로 사칙연산이 가능하며, 동일한 위치의 값끼리 연산이 이루어진다.
    - 두 개의 DataFrame의 길이가 달라서 연산 할 수 없는 경우 NaN으로 처리된다.    
    <br>
- Dictionary로 정의된 DataFrame의 연산 : Dictionary로 DataFrame을 정의하면 Dictionary의 Key 가 DataFrame의 Column 이 된다.
    - 동일한 이름의 Column을 기준으로 연산 된다. 만약 동일한 Column이 없으면?  --> 해당 열의 값은 전부 NaN
    - 문자열 Data의 경우 문자열이 이어붙여지는 방식으로 덧셈 연산이 이루어진다. 또한 문자열의 경우 덧셈이 아닌 다른 사칙연산을 하면 Error 발생

### Series의 접근법 (Series는 1차원 배열 형태이므로, index에 의한 접근만 있다. )
### 행(index)로 접근하는 방법
 - index : iloc 와 loc 이 있다. 
 - loc : index 값(내용) 으로 접근 
 - loc 을 활용한 슬라이싱 : 행 단위로, 범위를 지정하여 선택
    - s1.loc[시작인덱스 : 끝 인덱스 : 간격]
 - iloc : index위치값(0 , 1, 2, 3...) 으로 접근  -  index는 0 부터 시작   
 - iloc 을 활용한 슬라이싱 : 행 단위로, 범위를 지정하여 선택 하는 방법
 - s1.iloc[시작인덱스 : 끝 인덱스 : 간격]   : 간격의 기본값은 1
    - ex) s1.iloc[::2]  ==> Series s1에 대해서 처음부터 2행 간격으로 선택

    - df.iloc[index의 순번 , column의 순번]
    - ex) df.iloc[0,1:3]  ==>  index의 순서가 0번째(첫번째) 행 에서 column의 순서가 1~3(2번째와 3번째) 인 데이터

    - df.iloc[:,1]   두번째 열의 모든 데이터
    - df.iloc[:,[1,3]]  두번째와 네번째 열의 모든 데이터

```python
import pandas as pd
from IPython.display import display

# Series데이터 's' 정의
s=pd.Series(range(-3,6))

# Series 데이터(s)에서 조건식을 적용해서 출력 
# print(s)
s[s>0] # 0보다 큰 값
s[s%2==0] # 짝수 값
s[(s>0) & (s%2==0)] # AND 조건 : 0 보다 크면서 짝수 
s[(s>0) | (s%2==0)] # OR 조건 : 0 보다 크거나 짝수 

#
```

### DataFrame Data에서 조건으로 데이터 선택
### 조건이 1개 일때 : DataFrame_data[조건식]
### 조건이 여래개 일때 : 
    AND조건 : DataFrame_data[(조건식) & (조건식)]
    OR 조건 : DataFrame_data[(조건식) | (조건식)]

### isin() : DataFrame에서 OR 조건과 같은 결과를 가져온다. 
    df[df['지점'].isin(['서울','부산'])]  :  데이터 프레임 'df'의 '지점'이 '서울' 이거나 '부산' 인 데이터

```python
import pandas as pd
from IPython.display import display

# dataframe을 만들 Dictionary 
dic_data={'지점':['서울','대전','대구','부산','광주'],'1월':[558,234,340,300,213],'2월':[437,216,238,290,194],'3월':[337,196,209,272,186]}

# Dictionary를 이용해서 DataFrame 'df'를 정의
df=pd.DataFrame(dic_data)

display(df)

# DataFrame 'df'에서 '1월' 필드의 값이 300 이상인 데이터
df[df['1월']>=300]

# 1월과 3월이 모두 300 이상인 데이터 - AND
df[(df['1월']>=300) & (df['3월']>=300)]

# 1월과 300 이상이거나 3월이 300 이상인 데이터 - OR
df[(df['1월']>=300) | (df['3월']>=300)]

#  데이터 프레임 'df'의 '지점'이 '서울' 이거나 '부산' 인 데이터
df[df['지점'].isin(['서울','부산'])] 


```
